```python
heritage = pd.read_excel('heritage-data-2019.xls')
heritage.head()
```

```python
def get_rank_score(name):
    result = heritage.loc[heritage['Country Name'] == name]
    if len(result) == 0:
        return (None,None)
    else:
        rank = result.iloc[0]['World Rank']
        score = result.iloc[0]['2019 Score']
        return (rank, score)
    
for index, name in df['Country (according to CN Gov):'].items():
    rank, score = get_rank_score(name)
    if rank != None:
        print(name, rank,score)
        df['']
    else:
        print('no data for',name)
```
